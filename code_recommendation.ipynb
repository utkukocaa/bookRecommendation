## Pearson coefficient değerlerinin hesaplanması için yazılan fonksiyon

```python
import math

def pearson_co(a,b):
    x = df_ru.loc[a].copy()
    y = df.loc[b].copy()

    def cleaned_array(a):   
        total = 0
        b = 0
        for i in range(a.size):
            if(a[i] != 0):
                total = total + a[i]
                b = b + 1
        mean = total / b
    
        for i in range(a.size):
            if(a[i] != 0):
                a[i] = a[i] - mean
        return a


    def cal_deno(x):
        for i in range(x.size):
            x[i] = x[i] * x[i]
        return math.sqrt(sum(x))



    v = cleaned_array(x)
    w = cleaned_array(y)
    nominator =  sum(v*w)
    denominator = cal_deno(v)*cal_deno(w)
    result = round((nominator / denominator), 3)
    return result
```

## Her hedef kullanıcı için benzerliğine bakılacak olan 20 kullanıcının pearson coefficient değerlerinin bir arada hesabı 

```python
  r = []
def hesapla(x):
    r.clear()
    list=["U1","U2","U3","U4","U5","U6","U7","U8","U9","U10","U11","U12","U13","U14","U15","U16","U17","U18","U19","U20"]
    for i in list:
        r.append(pearson_co(x,i))
    return r.copy()
    
```
